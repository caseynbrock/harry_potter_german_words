```python
%matplotlib inline
import numpy as np
import mycount
import pandas as pd
import matplotlib.pyplot as plt
```

```python
formatted_harry_txt = mycount.format_harry_potter('harry_potter_stein_weisen.txt')  # read and format harry potter text
kw = mycount.read_known_word_list('duolingo_words_learned.txt')  # read list of known words
b = mycount.Book(formatted_harry_txt, kw)  # create Book object with harry potter data
df=b.word_data  # data frame of word statistics
pd.set_option('display.max_rows', 200)   # setting to print 200 lines of Pandas tables
```

How many new words are there?

```python
print b.num_unique_words - len(b.known_words), ' new words'
```

How many words appear only once in the book?

```python
print len(df[df['word_count']==1]), ' words appear only once'
```

```python
print len(df[(df['word_count']>5) & (df['learned']==False)]) # number of unlearned words that appear more than 5 times in book
```

```python
known_word_probability = df['word_count']/b.total_words
ax= plt.plot(np.cumsum(known_word_probability)*100)
plt.axvline(x=len(b.known_words), color='k')
plt.text(len(b.known_words)+50, 45, 'duolingo words')
plt.xlabel('number of most popular words learned')
plt.ylabel('percent of Harry Potter words covered')
plt.xlim([0,b.num_unique_words])
plt.ylim([0,100])
plt.grid()
plt.show()
```

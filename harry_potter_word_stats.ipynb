```python
import mycount
import pandas as pd
```

```python
formatted_harry_txt = mycount.format_harry_potter('harry_potter_stein_weisen.txt')
kw = mycount.read_known_word_list('duolingo_words_learned.txt')
b = mycount.Book(formatted_harry_txt, kw)
df=b.word_data
pd.set_option('display.max_rows', 200)   # to print full pandas tables
```

How many new words are there?

```python
print b.num_unique_words - len(b.known_words), ' new words'
```

How many words appear only once in the book?

```python
print len(df[df['count']==1]), ' words appear only once'
```

Print first 200 new words that appear more than five times in book, sorted by page on which they first appear. The total number of occurences of each word in the book are listed in the "count" column.

```python
chunk = df[(df['count']>5) & (df['learned']==False)].sort_values(['page']).iloc[0:200]
chunk
```

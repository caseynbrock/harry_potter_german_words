```python
formatted_harry_txt = mycount.format_harry_potter('harry_potter_stein_weisen.txt')
kw = mycount.read_known_word_list('duolingo_words_learned.txt')
b = mycount.Book(formatted_harry_txt, kw)
df=b.word_data
```

Print all words that appear more than five times in book and weren't learned on duolingo, sorted by page on which they first appear:

```python
chunk = df[(df['count']>5) & (df['learned']==False)].sort_values(['page'])
chunk
```
